```python
import pandas as pd
import bar_chart_race as bcr
```

```python
df = pd.read_csv("file.csv")
df.head()
```

```python
df.rename(columns={'Meat, total | 00001765 || Production | 005510 || tonnes': 'Meat production (ton)'}, inplace=True)
teste = df.pivot_table(index=['Year'], columns=['Country'], values=['Meat production (ton)'], aggfunc='sum')
teste.fillna(0, inplace=True)
teste = teste.cumsum(axis=0)
teste = teste['Meat production (ton)'][['Africa (FAO)', 'Americas (FAO)', 'Europe (FAO)', 'Asia (FAO)', 'Oceania (FAO)']]
teste.head()
```

```python
bcr.bar_chart_race(teste, figsize=(4, 2.5), title='Produção de carne', n_bars=10, filename='teste.mp4')
```
